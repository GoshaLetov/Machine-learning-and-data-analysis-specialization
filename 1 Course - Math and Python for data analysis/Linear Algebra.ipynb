## Задача №1    Сравнение предложений

```python
import re
import warnings

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.spatial.distance import cosine
from scipy.linalg import solve

%matplotlib inline

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (12, 8)
warnings.filterwarnings('ignore')
```

```python
"""Собираем предложения в список"""
sentences = list()
with open('Data/sentences.txt', 'r') as file:
    for line in file:
        sentences.append(line.lower()[:-2])
    file.close()
```

```python
"""Собираем слова в список"""
words = list()
for sentence in sentences:
    for word in re.split('[^a-z]', sentence): 
        if word != '':
            words.append(word)
```

```python
"""Создаем словарь индесов слов"""
words_dict = dict(zip(set(words), [i for i in range(len(set(words)))]))
```

```python
"""Собираем матрицу вхождения слов"""
token_matrix = np.zeros([len(words_dict), len(sentences)])
for idx, sentence in enumerate(sentences):
    bag_of_words = [word for word in re.split('[^a-z]', sentence) if word != '']
    container = list()
    for word, value in zip(words_dict.keys(), words_dict.values()):
        token_matrix[value, idx] = bag_of_words.count(word)
```

```python
cosine_dists = list()
for idx in range(token_matrix.shape[1]):
    cosine_dists.append([idx, cosine(token_matrix[:, 0], token_matrix[:, idx])])
```

```python
cosine_dists = pd.DataFrame(
    data=[i[1] for i in cosine_dists],
    columns=['cosine_dists'],
    index=[i[0] for i in cosine_dists]
)
```

```python
print (f'Первое предложение: \n{sentences[0]}\n')
for idx in cosine_dists.sort_values('cosine_dists', ascending=False).index[:2]:
    print (f'Номер предложения = {idx}, косинусное расстояние = {cosine_dists.loc[idx, "cosine_dists"]}')
    print (f'Предложение: \n{sentences[idx]}\n')
```

## Задача №2 Аппроксимация функции

```python
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

```python
plt.plot(range(1, 15), [f(x) for x in range(1, 15)])
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('График функции')
```

```python
def slau_creator(dots_array):
    b = np.array([f(x) for x in dots_array])
    A = np.ones([len(dots_array), len(dots_array)])
    for idx in range(A.shape[0]):
        A[idx, :] = dots_array[idx]
        for col in range(A.shape[1]):
            A[idx, col] = A[idx, col] ** col
    return A, b
```

```python
def approx_f(w, x):
    x_n = [x ** i for i in range(w.shape[0])]
    return np.dot(w, x_n)
```

```python
def approx_f_plot(dots_array):
    A, b = slau_creator(dots_array)
    slv = solve(A, b)
    x = dots_array
    y = [approx_f(slv, i) for i in x]
    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Аппроксимация функции f(x)')
    plt.show()
```

```python
approx_f_plot(range(1, 15))
```

```python
approx_f_plot([1, 8, 15])
```

```python
approx_f_plot([1, 4, 10, 15])
```
